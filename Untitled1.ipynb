```python
import os
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis, entropy
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.optimize import curve_fit
from sklearn.metrics import mean_squared_error
```

```python
data_folder = './1st_test/1st_test'  # update this path as needed

# Get sorted list of file names
file_list = sorted(os.listdir(data_folder))
total_files = len(file_list)
print("Total files:", total_files)

# Create RUL labels (assuming file order is time-ordered)
rul_labels = np.arange(total_files - 1, -1, -1)
```

```python
def mean_abs(signal):
    return np.mean(np.abs(signal))

def std_dev(signal):
    return np.std(signal)

def skewness(signal):
    return skew(signal)

def kurt(signal):
    return kurtosis(signal)

def shannon_entropy(signal, bins=500):
    hist, _ = np.histogram(signal, bins=bins, density=True)
    hist = hist[hist > 0]
    return entropy(hist)

def rms(signal):
    return np.sqrt(np.mean(signal**2))

def peak_to_peak(signal):
    return np.max(signal) - np.min(signal)

def crest_factor(signal):
    r = rms(signal)
    return np.max(np.abs(signal)) / r if r != 0 else 0

def clearance_factor(signal):
    denominator = (np.mean(np.sqrt(np.abs(signal))))**2
    return np.max(np.abs(signal)) / denominator if denominator != 0 else 0

def shape_factor(signal):
    m_abs = np.mean(np.abs(signal))
    return rms(signal) / m_abs if m_abs != 0 else 0

def impulse_factor(signal):
    m_abs = np.mean(np.abs(signal))
    return np.max(np.abs(signal)) / m_abs if m_abs != 0 else 0
```

```python
features_list = []

for fname in file_list:
    file_path = os.path.join(data_folder, fname)
    # Load file (each file is a 2D array: 20480 x 8)
    data = np.loadtxt(file_path)
    file_features = {}
    # Loop over channels (assuming 8 channels)
    for ch in range(data.shape[1]):
        signal = data[:, ch]
        prefix = f'ch{ch+1}'
        file_features[f'{prefix}_mean_abs'] = mean_abs(signal)
        file_features[f'{prefix}_std'] = std_dev(signal)
        file_features[f'{prefix}_skew'] = skewness(signal)
        file_features[f'{prefix}_kurtosis'] = kurt(signal)
        file_features[f'{prefix}_entropy'] = shannon_entropy(signal)
        file_features[f'{prefix}_rms'] = rms(signal)
        file_features[f'{prefix}_p2p'] = peak_to_peak(signal)
        file_features[f'{prefix}_crest'] = crest_factor(signal)
        file_features[f'{prefix}_clearance'] = clearance_factor(signal)
        file_features[f'{prefix}_shape'] = shape_factor(signal)
        file_features[f'{prefix}_impulse'] = impulse_factor(signal)
    features_list.append(file_features)

df_features = pd.DataFrame(features_list)
df_features['RUL'] = rul_labels  # Add RUL label
print("Feature DataFrame shape:", df_features.shape)
df_features.head()
```

```python
df_smoothed = df_features.copy()
feature_cols = df_smoothed.columns.drop('RUL')

for col in feature_cols:
    df_smoothed[col] = df_smoothed[col].ewm(span=40, adjust=False).mean()

df_smoothed.head()
```

```python
X = df_smoothed.drop(columns=['RUL'])
pca = PCA()
X_pca = pca.fit_transform(X)
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])

# Check explained variance of PC1
print(f"Explained variance of PC1: {pca.explained_variance_ratio_[0]:.4f}")

# For this example, assume that a rising PC1 indicates degradation.
# If PC1 is decreasing instead, flip it:
if df_pca['PC1'].iloc[-1] < df_pca['PC1'].iloc[0]:
    df_pca['PC1'] = df_pca['PC1'].max() - df_pca['PC1']

# Build a DataFrame for degradation modeling
df_health = pd.DataFrame({
    'PC1': df_pca['PC1'] - df_pca['PC1'].min(),  # normalize to start at 0
    'cycle': np.arange(len(df_pca)),
    'RUL': df_smoothed['RUL'].values
})
df_health.head()
```

```python
def exp_func(x, a, b):
    return a * np.exp(b * x)

# Choose a window (e.g., last 250 cycles)
fit_window = 250
x_fit = df_health['cycle'].values[-fit_window:]
y_fit = df_health['PC1'].values[-fit_window:]

# Fit the exponential curve
params, _ = curve_fit(exp_func, x_fit, y_fit, p0=(1, 0.001), maxfev=10000)
a_fit, b_fit = params
print(f"Fitted parameters: a = {a_fit:.5f}, b = {b_fit:.5f}")

# Generate fitted values for visualization
y_fitted = exp_func(x_fit, a_fit, b_fit)
```

```python
# Set a failure threshold (you can adjust this based on your domain knowledge)
threshold = 3.0  # Example threshold; adjust as needed

# Compute predicted failure cycle
predicted_failure_cycle = np.log(threshold / a_fit) / b_fit
current_cycle = df_health['cycle'].iloc[-1]
predicted_rul = predicted_failure_cycle - current_cycle

print(f"Predicted failure at cycle: {predicted_failure_cycle:.0f}")
print(f"Predicted RUL from current cycle {current_cycle}: {predicted_rul:.0f}")
```

```python
# For a simple evaluation, we assume the final cycle (failure) is when RUL=0.
final_cycle = df_health['cycle'].max()

# Create a DataFrame with evaluation points; here we use the predictions made at the current cycle
df_eval = pd.DataFrame({
    'cycle': [current_cycle],
    'predicted_rul': [predicted_rul],
    'true_rul': [final_cycle - current_cycle]
})
print(df_eval)

# Compute RMSE for this single prediction (for demonstration)
rmse = np.sqrt(mean_squared_error(df_eval['true_rul'], df_eval['predicted_rul']))
print(f"RMSE: {rmse:.2f}")
```

```python
log = {'cycle': [], 'predicted_rul': []}
increment = 25  # update every 25 cycles

for current_cycle in range(600, final_cycle, increment):
    segment = df_health.iloc[:current_cycle]
    # Use the last fit_window cycles; if current_cycle is less than fit_window, use all available cycles.
    if current_cycle < fit_window:
        continue
    x_seg = segment['cycle'].values[-fit_window:]
    y_seg = segment['PC1'].values[-fit_window:]
    
    try:
        params_seg, _ = curve_fit(exp_func, x_seg, y_seg, p0=(1, 0.001), maxfev=10000)
        a_seg, b_seg = params_seg
        predicted_fail_cycle = np.log(threshold / a_seg) / b_seg
        pred_rul = predicted_fail_cycle - x_seg[-1]
        log['cycle'].append(x_seg[-1])
        log['predicted_rul'].append(pred_rul)
    except RuntimeError:
        # Skip if fit fails
        continue

df_online = pd.DataFrame(log)
plt.figure(figsize=(10, 6))
plt.plot(df_online['cycle'], df_online['predicted_rul'], marker='o', label='Predicted RUL')
plt.axhline(0, color='gray', linestyle='--', label='Failure')
plt.xlabel("Cycle")
plt.ylabel("Predicted RUL")
plt.title("Online Predicted RUL Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
