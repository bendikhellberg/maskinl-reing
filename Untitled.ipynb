```python
import numpy as np

total_files = 2156
rul_labels = np.arange(total_files - 1, -1, -1)  # 2155, 2154, ..., 0
```

```python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis, entropy

# Define each feature function
def calculate_rms(signal): 
    return np.sqrt(np.mean(signal**2))

def calculate_p2p(signal): 
    return np.max(signal) - np.min(signal)

def calculate_entropy(signal): 
    hist, _ = np.histogram(signal, bins=500, density=True)
    hist = hist[hist > 0]
    return entropy(hist)

def calculate_crest(signal): 
    return np.max(np.abs(signal)) / calculate_rms(signal)

def calculate_clearance(signal): 
    return np.max(np.abs(signal)) / ((np.mean(np.sqrt(np.abs(signal))))**2)

def calculate_shape(signal): 
    return calculate_rms(signal) / np.mean(np.abs(signal))

def calculate_impulse(signal): 
    return np.max(np.abs(signal)) / np.mean(np.abs(signal))
```

```python
import os
from tqdm import tqdm

# Initialize
feature_list = []

# Loop through all files
folder_path = './1st_test/1st_test'
file_list = sorted(os.listdir(folder_path))  # ensure time order

for fname in tqdm(file_list):
    file_path = os.path.join(folder_path, fname)
    data = np.loadtxt(file_path)  # shape: (20480, 8)

    features = {}

    for ch in range(data.shape[1]):  # Loop over 8 channels
        signal = data[:, ch]
        prefix = f'ch{ch+1}'

        features[f'{prefix}_mean'] = np.mean(np.abs(signal))
        features[f'{prefix}_std'] = np.std(signal)
        features[f'{prefix}_skew'] = skew(signal)
        features[f'{prefix}_kurtosis'] = kurtosis(signal)
        features[f'{prefix}_entropy'] = calculate_entropy(signal)
        features[f'{prefix}_rms'] = calculate_rms(signal)
        features[f'{prefix}_max'] = np.max(np.abs(signal))
        features[f'{prefix}_p2p'] = calculate_p2p(signal)
        features[f'{prefix}_crest'] = calculate_crest(signal)
        features[f'{prefix}_clearance'] = calculate_clearance(signal)
        features[f'{prefix}_shape'] = calculate_shape(signal)
        features[f'{prefix}_impulse'] = calculate_impulse(signal)

    feature_list.append(features)

# Convert to DataFrame
df_features = pd.DataFrame(feature_list)

# Add RUL labels
df_features['RUL'] = rul_labels

# Preview
print(df_features.shape)
print(df_features.head())
```

```python
df_ema = df_features.copy()

# Get all feature columns (exclude RUL)
feature_cols = df_ema.columns.drop('RUL')

# Apply EMA smoothing with span=40
for col in feature_cols:
    df_ema[col] = df_ema[col].ewm(span=40, adjust=False).mean()

# Check result
print(df_ema.head())
```

```python
from sklearn.decomposition import PCA

# Drop RUL for PCA
X_ema = df_ema.drop(columns=['RUL'])

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_ema)

# Store principal components in DataFrame
df_pca = pd.DataFrame(X_pca, columns=[f"PC{i+1}" for i in range(X_pca.shape[1])])

# Show how much variance PC1 explains
print(f"Explained variance of PC1: {pca.explained_variance_ratio_[0]:.4f}")

# Append PC1 and RUL for degradation modeling
df_health = pd.DataFrame({
    'PC1': df_pca['PC1'] - df_pca['PC1'].min(),  # normalized to start at 0
    'cycle': np.arange(len(df_pca)),
    'RUL': df_ema['RUL'].values
})

# Preview
print(df_health.head())
```

```python
from scipy.optimize import curve_fit
import numpy as np

# Define the exponential function
def exp_fit(x, a, b):
    return a * np.exp(b * x)

# Number of latest cycles to use for fitting
fit_window = 250

# Get the last `fit_window` points
x_data = df_health['cycle'][-fit_window:].values
y_data = df_health['PC1'][-fit_window:].values

# Fit the exponential curve
params, _ = curve_fit(exp_fit, x_data, y_data, p0=(0.01, 0.001), maxfev=10000)

# Predict the fitted curve on current x
y_fit = exp_fit(x_data, *params)

# Display fitted parameters
print(f"Fitted parameters: a = {params[0]:.5f}, b = {params[1]:.5f}")
```

```python
# Threshold at which we define failure
threshold = 2.0

# Extract a, b from fitted parameters
a, b = params

# Predict failure cycle
predicted_failure_cycle = np.log(threshold / a) / b

# Get current cycle (last one from your data)
current_cycle = df_health['cycle'].iloc[-1]

# Calculate predicted RUL
predicted_rul = predicted_failure_cycle - current_cycle

print(f"Predicted failure at cycle: {predicted_failure_cycle:.0f}")
print(f"Predicted RUL from current cycle {current_cycle}: {predicted_rul:.0f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df_health['cycle'], df_health['PC1'], label='PC1 (Health Index)', alpha=0.8)
plt.plot(x_data, y_fit, 'r--', label='Exponential Fit')
plt.axhline(threshold, color='gray', linestyle='--', label='Failure Threshold')
plt.xlabel('Cycle')
plt.ylabel('PC1')
plt.title('Exponential Degradation Curve')
plt.legend()
plt.grid()
plt.show()
```

```python
log = {
    'cycle': [],
    'predicted_failure_cycle': [],
    'predicted_rul': []
}

prediction_start = 600
increment = 25
fit_window = 250
threshold = 2.0

for cycle in range(prediction_start, len(df_health), increment):
    segment = df_health.iloc[:cycle]
    x = segment['cycle'].values[-fit_window:]
    y = segment['PC1'].values[-fit_window:]

    try:
        # Fit curve
        params, _ = curve_fit(exp_fit, x, y, p0=(0.01, 0.001), maxfev=10000)
        a, b = params

        # Predict failure
        predicted_cycle = np.log(threshold / a) / b
        predicted_rul = predicted_cycle - x[-1]

        # Save
        log['cycle'].append(x[-1])
        log['predicted_failure_cycle'].append(predicted_cycle)
        log['predicted_rul'].append(predicted_rul)
    except:
        continue  # Skip failed fits

# Convert to DataFrame
df_predictions = pd.DataFrame(log)
print(df_predictions.tail())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_predictions['cycle'], df_predictions['predicted_rul'], marker='o', linestyle='-', label='Predicted RUL')
plt.axhline(y=0, color='gray', linestyle='--', label='Failure Line')
plt.xlabel('Cycle')
plt.ylabel('Predicted RUL')
plt.title('Predicted RUL Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Get max cycle (final cycle before failure)
final_cycle = df_health['cycle'].max()

# Merge predictions with true RUL
df_eval = df_predictions.copy()
df_eval['true_rul'] = final_cycle - df_eval['cycle']

# Filter out invalid RULs (negative or extreme)
df_eval_filtered = df_eval[df_eval['predicted_rul'].between(0, 2 * final_cycle)]

# Compute RMSE
rmse = np.sqrt(mean_squared_error(df_eval_filtered['true_rul'], df_eval_filtered['predicted_rul']))
print(f"üìè RMSE between predicted and true RUL: {rmse:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_eval_filtered['cycle'], df_eval_filtered['true_rul'], label='True RUL', linestyle='--')
plt.plot(df_eval_filtered['cycle'], df_eval_filtered['predicted_rul'], label='Predicted RUL', marker='o')
plt.xlabel('Cycle')
plt.ylabel('RUL')
plt.title('True vs Predicted RUL Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
